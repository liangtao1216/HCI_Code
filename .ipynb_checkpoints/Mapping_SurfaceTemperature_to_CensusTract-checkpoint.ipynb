```python
import pandas as pd
import geopandas as gpd
import rasterio
import rasterstats
from rasterio.plot import show
from rasterio.plot import show_hist
from matplotlib import pyplot as plt
```

```python
#read the census tract
```

```python
ct2010 = gpd.read_file('./Data/Processed_Data/ct2010_geoid.geojson', driver='GeoJSON')
ct2010.head()
```

```python
ct2010.plot();
plt.show()
```

```python
# read surface temperature raster data of 20190830
```

```python
st = rasterio.open('./Data/Raw_Data/Environment/LC08_CU_029007_20190830_20190919_C01_V01_ST/LC08_CU_029007_20190830_20190919_C01_V01_ST.tif'
                   , mode='r')
```

```python
show(st)
```

```python
# project census tract to same coordinate system
ct2010 = ct2010.to_crs(st.crs)
```

```python
st.crs
```

```python
# plotting the raster and the census tract together
fig, (ax1, ax2) = plt.subplots(1,2, figsize=(10,4))
show(st,ax=ax1, title = 'surface temperature')
ct2010.plot(ax=ax1, facecolor='None', edgecolor='yellow')

show_hist(st, title='Histogram',ax=ax2) #Kelvins temperature (*10)
plt.show()
```

```python
type(st)
```

```python
# assign raster values to a numpy nd array
```

```python
st_array = st.read(1)
```

```python
st_array
```

```python
st.meta
```

```python
# Affine(30.0(width of each pixel), 0.0(rotation), 1784415.0(x coordinate of left upper corner), (column rotation) 
#        ,(heigh of the pixel), (y coordinate of left upper corner))
```

```python
st.nodata
```

```python
st.transform
```

```python
affine = st.transform
```

```python
# calculating the zonal statistics
```

```python
avg_st = rasterstats.zonal_stats(ct2010,st_array, affine=affine
                                 , stats=['mean'] #max min
                                , geojson_out = True)
```

```python
avg_st
```

```python
avg_st[0]['properties']
```

```python
# extracting the average surface temperature from the list
```

```python
avg_surfaceT = []
i = 0
while i< len(avg_st):
    avg_surfaceT.append(avg_st[i]['properties'])
    i = i+1
```

```python
# transfering the list to pandas dataframe
avg_surfaceT = pd.DataFrame(avg_surfaceT)
avg_surfaceT.head()
```

```python
# Convert 10*Kelvin temperature to Fahrenheit
avg_surfaceT['SurfaceT_F'] = 9*(avg_surfaceT['mean']/10-273.15)/5+32
```

```python
avg_surfaceT_geo = avg_surfaceT.merge(ct2010[['GeoID','geometry']],on='GeoID')
```

```python
avg_surfaceT_geo = gpd.GeoDataFrame(avg_surfaceT_geo,geometry=avg_surfaceT_geo['geometry'], crs=st.crs)
```

```python
avg_surfaceT_geo.crs
```

```python
avg_surfaceT_geo = avg_surfaceT_geo.to_crs('epsg:4326')
```

```python
avg_surfaceT_geo.plot(column='SurfaceT_F', cmap='Spectral_r', 
                              legend=True, scheme='Quantiles', k=10,
                              figsize=(10,10))
plt.show()
```
